```python
wine_reviews = pd.read_csv("/content/winemag-data-130k-v2.csv", index_col = 0)
wine_reviews
```

```python
#create a fig and axis
fig, ax = plt.subplots()

#count occurence of each cell
data = wine_reviews['points'].value_counts()

#get values of x and y data
points=data.index
frequency= data.values

#create bar charts
ax.bar(points, frequency)

#set title and labels
ax.set_title(" Wine Reviews Score")
ax.set_xlabel("Points")
ax.set_ylabel("Frequency")


wine_reviews['points'].value_counts().sort_index().plot.bar()
```

```python
wine_reviews.groupby('country').price.mean().sort_values(ascending = False)[:6].plot.bar()
```

```python
objects = ("Python", "C++", "Java", "Perl", "Scala", "Lisp")
y_posi= np.arange(len(objects))

performance = [10,8,6,9,5,3]

plt.bar(y_posi, performance, width = 0.5, align = 'center', alpha = 0.7, color = ['r', 'b', 'pink', 'g', 'y', 'orange'])
plt.xlabel('Programming Languages')
plt.ylabel('Usage')
plt.xticks(y_posi, objects)
plt.title('Programming language usage')
```

```python
# Declaring the figure or the plot (y, x) or (width, height)
plt.figure(figsize = (12, 7))

# Categorical data: Country names
countries = ['USA', 'Brazil', 'Russia', 'Spain', 'UK', 'India']

# integer value interms of death counts
totalDeaths = [112596, 37312, 5971, 27136, 40597, 7449]

# Passing the parameters to the bar function, this is the main function which creates the bar plot
plt.bar(countries, totalDeaths, width = 0.9, align = 'center', color = 'blue', edgecolor = 'red')

# This is the location for the annotated text
i = 1.0
j = 2000

# Annotating the bar plot with the values (total death counts)
for i in range(len(countries)):
    plt.annotate(totalDeaths[i], (-0.1 + i, totalDeaths[i] + j))

# Creating legend of the bars in the plot
plt.legend(labels = ['Total Deaths'])

# Giving the title for the plot
plt.title('Bar plot representing the total deaths by top 6 countries due to coronavirus')

# Naming the x and y axis
plt.xlabel('Countries')
plt.ylabel('Deaths')

# Saving the plot as a 'png'
plt.savefig("BarPlot.png")

# Display the bar plot
plt.show()
```

#**Stacking Barplot**

```python
df = pd.DataFrame({
    'name': ["John","mary",'peter','jeff','bill','lisa','jose'],
    'age':[23,40,70,20,35,65,44],
    'gender':['M','F','M','M','M','F','M'],
    'state':['California','dc','California','dc','California','texas','texas'],
    'num_children':[2,0,0,3,2,1,4],
    'num_pets':[5,1,0,5,2,2,3]
})

#from pandas to plot multiple plots on same figure
df.groupby(['state','gender']).size().unstack().plot(kind='bar',stacked=True)
```

```python
df.groupby(['gender','state']).size().unstack().plot(kind='bar',stacked=True)
```

```python
plt.figure(figsize=[15,10])

totalDeath = [11562, 4500, 3241, 11023, 9520]
totalRecovery = [234621, 201345, 102101, 98564, 103543]
activeCases = [112456, 22349, 34221, 11252, 35421]
country = ['India', 'Italy', 'Peru', 'Germany', "Iran"]

x = np.arange(len(totalDeath))

plt.bar(x, totalDeath, color ='red',width = 0.25)
plt.bar(x + 0.25, totalRecovery, color ='b',width = 0.25)
plt.bar(x + 0.5, activeCases, color ='g',width = 0.25)

plt.legend(['Total Death','Total Recovery','Active Cases'])

plt.xticks([i + 0.25 for i in range(5)], country)
plt.title('Bar plot representing Total Death,Total Recovery and the Active cases')

plt.xlabel("Countries")
plt.ylabel("Number of Cases")

plt.savefig("4BarPlot.png")

plt.show()
```

#**Pie Chart**

```python
from pandas import DataFrame

Data = {"Tasks":[300,500,800],
     'Task Type':['Task pending','Tasks Ongoing','Tasks Completed']
     }

df = DataFrame(Data)
df.set_index('Task Type',inplace=True)

df.plot.pie(y ='Tasks', figsize = (5, 5), autopct = '%1.1f%%', startangle = 90)
```

```python
%matplotlib inline

labels = ['Civil','Electrical','Mechanical','Chemical']
sizes = [15,30,45,10]

explode = (0.1,0.1,0.1,0.4)

fig, ax = plt.subplots()
ax.pie(sizes,
      explode=explode,
      labels=labels,
      autopct='%1.1f%%',
       shadow=True,
       startangle=90)

ax.axis('equal')
ax.set_title("Engineering Discilplines")
```
